```python
import pandas as pd
```

```python
df=pd.read_csv('/content/datafile - Sheet1.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna(method='ffill') #forward fill
```

```python
df.fillna(method='ffill',axis=1) #if axis is 1 then its column wise filling , and if axis is 0 then its row wise filling
```

```python
df.fillna(method='bfill',inplace = True)
df
```
