```python
import pandas as pd
```

```python
df = pd.read_csv('/content/bank_transactions.csv')
df.head()
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

***DROPPING ROWS WITH NULL VALUES***

```python
df.shape
```

```python
df2 = df.dropna(axis = 0) #dropna will drop all the rows from the data set ; axis  = 0 is default it will drop rows and axis  = 1 it will drop columns
```

```python
df2.shape
```

***DROPPING COLUMNS WITH NULL VALUES***

```python
df3 = df.dropna(axis = 1)
```

```python
df3.shape
```

```python
df.dropna(how = 'any') #if any value is null then remove that row
```

```python
df.dropna(how = 'all') #if any value is null then remove that column
```

```python
df.dropna(inplace = True )
```
