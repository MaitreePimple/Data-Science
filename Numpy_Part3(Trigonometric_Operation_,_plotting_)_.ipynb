Using Matplotlib with numpy

```python
x = np.arange(1,11)
y = np.arange(10,110,10)

plt.figure(figsize=(10,5))
plt.plot(x,y,'b--')
plt.show()
```

Plotting Trignometric curves

```python
x_sin = np.arange(0,2*np.pi,0.1)
y_sin = np.sin(x_sin)

print(y_sin)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x_sin,y_sin)
plt.title('Sin Curve')
plt.show()
```

```python
x_cos = np.arange(0,2*np.pi,0.1)
y_cos = np.cos(x_cos)

print(y_cos)

plt.figure(figsize=(10,5))
plt.plot(x_cos,y_cos)
plt.title('Cos Curve')
plt.show()
```

```python
x_tan = np.arange(0,2*np.pi,0.1)
y_tan = np.tan(x_tan)

print(y_tan)

plt.figure(figsize=(10,5))
plt.plot(x_tan,y_tan)
plt.title('Tan Curve')
plt.show()
```

```python
x_cot = np.arange(0,2*np.pi,0.1)
y_cot = 9/np.tan(x_cot)

print(y_cot)

plt.figure(figsize=(10,5))
plt.plot(x_cot,y_cot)
plt.title('Cot Curve')
plt.show()
```

Sub Plot

```python
plt.figure(figsize=(10,5))
plt.subplot(2,2,1)
plt.plot(x_sin, y_sin,'b--')
plt.title('Sin Curve')

plt.subplot(2,2,2)
plt.plot(x_cos,y_cos,'b--')
plt.title('Cos Curve')

plt.subplot(2,2,3)
plt.plot(x_cot,y_cot,'b--')
plt.title('Cot Curve')

plt.subplot(2,2,4)
plt.plot(x_tan,y_tan,'b--')
plt.title('Tan Curve')

plt.show()
```
