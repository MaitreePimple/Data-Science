*GROUPBY*



```python
import pandas as pd
```

```python
df = pd.read_csv('/content/sample2.csv')
df.head()
```

```python
branch_group = df.groupby(by = 'Branch') # group by is going is to make group by the given parameters ('by' means parameter )
branch_group
```

```python
branch_group.groups
```

```python
df.groupby(by = ['Branch','Section'])
branch_group.groups #for gaining the proper answer do put full path link from your dektop not from colab
```

```python
for group, data_frame in branch_group:
  print(group)
  print(data_frame)
```

***Merge() Function ** :* it will merge the data frame*

```python
df1 = pd.DataFrame({'Roll No.':[1,2,3,4,5], 'Physics':[34,45,36,47,30]})
df1
```

```python
df2 = pd.DataFrame({'Roll No.':[1,2,3,4,5], 'Maths':[80,78,65,70,46]})
df2
```

```python
pd.merge(df1,df2, on = 'Roll No.')
```

```python
pd.merge(df2,df1, on = 'Roll No.')
```

```python
df3 = pd.DataFrame({'Roll No.':[1,2,3,4,5], 'History':[80,78,65,70,46]})
df4 = pd.DataFrame({'Roll No.':[1,2,6,7,8], 'Geo':[78,34,56,40,96]})
pd.merge(df3,df4)
```

```python
pd.merge(df3,df4, how = 'left')
```

```python
pd.merge(df3,df4, how = 'right')
```

```python
pd.merge(df3,df4, how = 'outer')
```
