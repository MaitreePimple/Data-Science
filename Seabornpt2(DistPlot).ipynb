

1.   histogram represent the data distribution by forming bins along the ramge of the data and then
2.   drawing bars to show the number of observation that fall in each bin




```python
df.describe()
```

```python
bins =[2,3,4,5,6,7,8,9,10]
sns.distplot(df['time_spend_company'],bins = bins)
plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'],bins = bins, kde = False)
```

```python
sns.distplot(df['time_spend_company'],bins = bins, hist = False)
```

```python
sns.distplot(df['time_spend_company'],bins = bins, rug = True ) #rug gives more clear data
```

```python
sns.distplot(df['time_spend_company'],bins = bins, rug = True )
plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'], bins=bins, rug=True,
hist_kws={'color': 'red', 'edgecolor': 'blue', 'linewidth': 3, 'alpha': 0.5})
plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'], bins=bins, rug=True,
hist_kws={'color': 'red', 'edgecolor': 'blue', 'linewidth': 3, 'alpha': 0.5},
             kde_kws={'color': 'orange', 'linewidth':3})
plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'], bins=bins, rug=True,
             color='blue',
#hist_kws={'color': 'red', 'edgecolor': 'blue', 'linewidth': 3, 'alpha': 0.5},
             #kde_kws={'color': 'orange', 'linewidth':3}
             )
plt.xticks(bins)
```

```python
sns.distplot(df['time_spend_company'], bins=bins, rug=True,
             color='blue', vertical ='True',
#hist_kws={'color': 'red', 'edgecolor': 'blue', 'linewidth': 3, 'alpha': 0.5},
             #kde_kws={'color': 'orange', 'linewidth':3}
             )
plt.xticks(bins)
```
