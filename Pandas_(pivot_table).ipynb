```python
import pandas as pd
```

```python
df = pd.read_csv('/content/sample2.csv')
df.head()
```

```python
pd.pivot_table(df, index='Branch',  aggfunc='count')
```

```python
pd.pivot_table(df, index='Branch',  aggfunc='sum')
```

```python
pd.pivot_table(df, index='Branch',  aggfunc='max') #maximum marks
```

```python
pd.pivot_table(df, index='Branch',  aggfunc='min') #minimum marks
```

```python
# Assuming 'df' is your DataFrame
pd.pivot_table(df, index='Branch',  columns='Section', aggfunc='count')
```
